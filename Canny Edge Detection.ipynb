```python
import cv2
import matplotlib.pyplot as plt # type: ignore
```

```python
img =cv2.imread(r'C:\Users\himan\Pictures\astro.jpg')
imgRGB =cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray =cv2.cvtColor(imgRGB, cv2.COLOR_RGB2GRAY)
img_blur =cv2.GaussianBlur(img_gray, (3,3), 0) # 3x3 kernel, meaning the blur will be applied considering a neighborhood of 3x3 pixels.
plt.title("Original Image")
plt.imshow(imgRGB)
```

```python
edges =cv2.Canny(image= img_blur, threshold1=100, threshold2=200) #first threshold for the hysteresis procedure in the Canny edge detection
edges =cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)  #threshold2: This is the second threshold for the hysteresis procedure.
edges =cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)  
plt.imshow(edges)
```
