```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# used to detect corners or interest points in an image. It was introduced by Chris Harris
```

```python
img=cv2.imread(r'D:\Data centr\CV_img\corner.jfif')
img_gry=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_gry=np.float32(img_gry)
plt.imshow(img)
```

```python
corner_img=cv2.cornerHarris(img_gry,2,5,0.07)  # 2: The size of the neighborhood
# 5: The size of the Sobel kernel 
# 0.07: The parameter k in the Harris corner response function.

corner_img=cv2.dilate(corner_img,None)
img[corner_img>0.01*corner_img.max()]=[0,0,255]
# pixel values in corner_img that are greater than 1% of the maximum value in corner_img are set to True (1),
plt.imshow(img)
```
