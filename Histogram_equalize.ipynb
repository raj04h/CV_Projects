```python
import cv2 # to enhance the contrast of an image. The goal of histogram equalization is to transform the intensity values of an image.
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r"C:\Users\himan\Pictures\bc.png")
imggray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

```python
dst=cv2.equalizeHist(imggray)
dst=cv2.cvtColor(dst,cv2.COLOR_GRAY2BGR)
plt.imshow(dst) # image contrast increase
```
