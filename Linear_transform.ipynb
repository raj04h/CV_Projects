```python
import cv2 # linear transformation, effects like contrast stretching, brightness adjustment, or negative transformation. 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r'C:\Users\himan\Pictures\apple.png')
img_inv=255-img
```

```python
fig=plt.figure(figsize=(15,5)) # means the figure will have a width of 15 units and a height of 5 units.
ax=fig.add_subplot(121) # 121 indicates that there will be 1 row, 2 columns of subplots, and this subplot will be the first one.
plt.title('original',fontsize=18)
plt.imshow(img,cmap='gray') # cmap parameter is set to 'gray', which means the image will be displayed in grayscale
```

```python
ax=fig.add_subplot(122) # e 1 before 22 indicates that there will be 1 row of subplots, and the 22 indicates that this subplot will be the second one.
plt.imshow(img_inv,cmap='gray')
ax.set_title('Inverse',fontsize=18)
plt.title('inverse',fontsize=18)
```
