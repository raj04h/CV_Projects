```python
import cv2  #Log transformation is often used to enhance the darker regions of an image.
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread(r'C:\Users\himan\Pictures\apple.png',0)
fig=plt.figure(figsize=(15,5))
ax= fig.add_subplot(121)
plt.title('original')
plt.imshow(img,cmap='gray')
```

```python
log=0.6*(np.log(1+np.float32(img))) # logarithmic transformation on the image using the formula: 
ax= fig.add_subplot(122) #  input image is treated as a floating-point array before applying the logarithm. 
plt.imshow(log,cmap='gray')
plt.title('log_transf') # 0.6 can be adjusted to control the intensity of the enhancement. Larger values will enhance the darker regions more strongly, while smaller values will have a more subtle effect.
```
