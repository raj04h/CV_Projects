```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r"C:\Users\himan\Pictures\astro.jpg")
plt.imshow(img)
if img is None:
    print("could not load image: ", img)
    exit(0)
```

```python
blue,green,red=cv2.split(img)
cv2.imshow('Blue- Gray Scale',blue)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow('Green- Gray Scale',green)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow('Red- Gray Scale',red)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
