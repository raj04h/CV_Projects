```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r'D:\Data centr\CV_img\sudoko.png')
img_gry=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

```python
edges=cv2.Canny(img_gry,50,150,apertureSize=3)
lines=cv2.HoughLines(edges,1,np.pi/180,200)
# 1: The resolution parameter rho
# 200: The threshold parameter.
```

```python
for line in lines:
    rho,theta=line[0]
    a=np.cos(theta)
    b=np.sin(theta)
    x0=a*rho
    y0=b*rho
    x1=int(x0+1000*(-b)) # 1000: This is a scaling factor that is used to determine the length of the line segment
    y1=int(y0+1000*(a))
    x2=int(x0-1000*(-b))
    y2=int(y0-1000*(a))
    cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
plt.imshow(img)   
    
```
