```python
import random
import cv2
from matplotlib import pyplot as plt
```

```python
def add_noise(img):
    row,col=img.shape
    num_of_pixels=random.randint(300,10000)
    for i in range(num_of_pixels):
        y_cord=random.randint(0,row-1)
        x_cord=random.randint(0,col-1)
        img[y_cord][x_cord]=255
    num_of_pixels=random.randint(300,10000)
    for i in range(num_of_pixels):
        y_cord=random.randint(0,row-1)
        x_cord=random.randint(0,col-1)
        img[y_cord][x_cord]=0
 
    return img
```

```python
img=cv2.imread(r'C:\Users\himan\Pictures\Screenshot_20230628-183949-612.jpg')
plt.title(img)
plt.imshow(img)
plt.show()
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
img=cv2.imread(r'C:\Users\himan\Pictures\Screenshot_20230628-183949-612.jpg',cv2.IMREAD_GRAYSCALE)
cv2.imwrite('noisy.jpg',add_noise(img))
noise=cv2.imread('noisy.jpg')
plt.imshow(noise)
plt.show()
 
```

```python
median=cv2.medianBlur(noise,5) # 5 indicate size of filter i.e. kernel size
plt.imshow(median) # imshow function to display the image after applying the median filter
plt.show() # displays the image on the screen
```
