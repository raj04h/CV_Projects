```python
import cv2
from matplotlib import pyplot as plt
```

```python
img= cv2.imread(r'C:\Users\himan\Pictures\OIP.jpg') 
```

```python
name='OIP' #image name show
plt.title(name)
plt.imshow(img) #image show
```

```python
horizontal_flip=cv2.flip(img,1) #fliping horizontal
name='OIP_Horizontally'
plt.title(name)
plt.imshow(horizontal_flip)
```

```python
vertical_flip=cv2.flip(img,0) #flip vertical
name='OIP_vertically' #0 indicte starting coordinate
plt.title(name)
plt.imshow(vertical_flip)
```

```python
vert_hori_flip=cv2.flip(img,-1) # flip_horiz_vert
name='OIP_vertically_horizontally' # -1 for horizontal and vertical fliped coordinate
plt.title(name)
plt.imshow(vert_hori_flip)
```

```python
plt.imshow(img)
```

```python
contrast=cv2.bitwise_not(img) #bitwise_not- Arthomatic operation that convert 0 to 1 & 1 to 0
plt.imshow(contrast)
cv2.imwrite('contrast.jpg',contrast)
```
