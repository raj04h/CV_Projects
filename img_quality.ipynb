```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread(r'C:\Users\himan\Pictures\Screenshot_20230628-183949-612.jpg')
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title("Original")
plt.imshow(img_rgb) # reading the image
```

```python
contrast=cv2.bitwise_not(img) #bitwise_not- Arthomatic operation that convert 0 to 1 & 1 to 0
plt.imshow(contrast) # 0 - black, 1 - white
plt.title("contrast") 
cv2.imwrite('contrast.jpg',contrast) #contrast of image
```

```python
blur_img=cv2.medianBlur(img,13)
plt.imshow(blur_img)
plt.title("blur")
cv2.imwrite('blur.jpg',blur_img)
```
