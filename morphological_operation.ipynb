```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
# morphological operations used for tasks like noise reduction, object detection, and shape manipulation.
```

```python
img=cv2.imread(r"C:\Users\himan\Pictures\draWING1.png")
kernel=np.ones((5,5),np.uint8) # larger kernel size will result in more extensive erosion and dilation effects.
plt.title('original')
plt.imshow(img)
```

```python
erosion=cv2.erode(img,kernel) # remove pixel on boundary
plt.title('erosion')  erosion operation removes pixels on the boundary of objects in the image, effectively reducing the size of the objects
plt.imshow(erosion)
```

```python
dilation=cv2.dilate(img,kernel) #add pixel on boundary
plt.title('dilation')  # dilation operation, adds pixels to the boundary of objects in the image, effectively increasing the size of the objects
plt.imshow(dilation)
```
