```python
import cv2 #Power-law transformation, gamma adjustment, used to adjust the brightness and contrast of an image.
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread(r'C:\Users\himan\Pictures\apple.png',0)
fig=plt.figure(figsize=(15,5)) #plt.figure with a size of 15x5 inches
ax= fig.add_subplot(121)
plt.title('original')
plt.imshow(img,cmap='gray')
```

```python
img1=img/255.0 # by dividing it convert img into 0 to 1
img_power_law=cv2.pow(img1,1.8) # normalized image gama=1.8
ax= fig.add_subplot(122)
plt.imshow(img_power_law,cmap='gray')
ax.set_title('power_law_trans')
```
